```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
from sklearn import datasets
centers = [[1, 1], [-1, -1], [1, -1]]
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
 # select two (random) dimensions to make visualization easier
X2 = X[:, [3, 0]]
```

```python
est = KMeans(n_clusters=3)
fig = plt.figure(1)
#ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
est.fit(X2)
labels = est.labels_

plt.scatter(X2[:, 0], X2[:, 1], c=labels.astype(np.float))
means = est.cluster_centers_
plt.scatter(means[:, 0], means[:, 1], c=[0,1,2], marker="*", s=100, linewidths=0)
```
